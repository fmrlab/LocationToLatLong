```python
n_records = data.shape[0]
google_geolocator = GoogleV3()
#csvfile = open('fire_dept_kentucky.csv', 'w')
fieldnames = ['name', 'address', 'latitude', 'longitude']
#writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
#writer.writeheader()
n_records = 20
```

```python
i = 0
with open('fire_dept_kentucky.csv', 'w') as csvfile:
    fieldnames = ['name', 'address', 'latitude', 'longitude']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    while i < n_records:
        fire_dept_name = data['Fire dept name'][i]
        addr = data['HQ addr1'][i]
        city = data['HQ city'][i]
        hq_zip = data['HQ zip'][i]
        query_str = fire_dept_name
        if addr == addr:
            query_str = query_str + ", " + addr
        if city == city:
            query_str = query_str + ", " + city
        query_str = query_str + ", " + "KY"
        if hq_zip == hq_zip:
            query_str = query_str + " " + hq_zip
        try:
            location = google_geolocator.geocode(query_str)
            print(location)
            print(location.latitude, location.longitude)
            writer.writerow({'name': fire_dept_name, 'address': location, 'latitude': location.latitude, 'longitude':location.longitude})
            i = i+1
        except Exception as e:
            print(e)
            time.sleep(4)
```
